# Data Preprocessing
This notebook downloads the [Breast Cancer Wisconsin Dataset](https://www.kaggle.com/datasets/uciml/breast-cancer-wisconsin-data), removes unnecessary columns, splits the data into 20% testing and 80% training and saves it to two separate CSV files

```python
import os
import pandas as pd
from sklearn.model_selection import train_test_split
import kagglehub
```

### Download and basic processing
Here we:
- **download the dataset** from [kagglehub](https://www.kaggle.com/datasets/uciml/breast-cancer-wisconsin-data)
- get the **path of the data** and **load** it into a pandas **dataframe**
- print the head of the dataframe for a quick look to the raw data
- **map the diagnosis** into a _"target"_ column with 0 as _Malignant_ and 1 as _Benign_
- **drop useless columns** and separate the data into **"data"** and **"labels"** **dataframes**

```python
path = kagglehub.dataset_download("uciml/breast-cancer-wisconsin-data")
print("Path to dataset files:", path)

csv_path = os.path.join(path, "data.csv")
df = pd.read_csv(csv_path)
print(df.head())

df["target"] = df["diagnosis"].map({
    "M": 0,
    "B": 1
})

data = df.drop(columns=["id", "diagnosis", "target", "Unnamed: 32"])
labels = df["target"]
```

## Train-Test splitting
We split the data and labels keeping **80%** for **training** and **20%** for **testing**

We then print the shape of the two subset to visualize how much raw data is kept for each phase

```python
X_train, X_test, y_train, y_test = train_test_split(
    data,
    labels,
    test_size=0.2,
    random_state=42,
    stratify=labels
)

print("Training shape:", X_train.shape)
print("Testing shape:", X_test.shape)
```

## Saving
We save the **training** and **testing** **dataframes** by combining labels and data of each and saving them to **local csv files**

```python
train_df = X_train.copy()
train_df["target"] = y_train.values

train_df.to_csv("training_data.csv", index=False)


test_df = X_test.copy()
test_df["target"] = y_test.values

test_df.to_csv("testing_data.csv", index=False)
```
