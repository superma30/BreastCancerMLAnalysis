# Data Preprocessing
This notebook downloads the [Breast Cancer Wisconsin](https://www.kaggle.com/datasets/uciml/breast-cancer-wisconsin-data) dataset, removes unnecessary columns, splits the data into 20% testing and 80% training and saves it to two separate CSV files

```python
import os
import pandas as pd
from sklearn.model_selection import train_test_split
import kagglehub
```

```python
os.environ['http_proxy'] = "http://172.18.0.1:3128"
os.environ['https_proxy'] = "http://172.18.0.1:3128"
```

```python
path = kagglehub.dataset_download("uciml/breast-cancer-wisconsin-data")
print("Path to dataset files:", path)

csv_path = os.path.join(path, "data.csv")
df = pd.read_csv(csv_path)
print(df.head())
```

```python
df["target"] = df["diagnosis"].map({
    "M": 0,
    "B": 1
})

data = df.drop(columns=["id", "diagnosis", "target", "Unnamed: 32"])
labels = df["target"]

X_train, X_test, y_train, y_test = train_test_split(
    data,
    labels,
    test_size=0.2,
    random_state=42,
    stratify=labels
)

print("Training shape:", X_train.shape)
print("Testing shape:", X_test.shape)
```

```python
train_df = X_train.copy()
train_df["target"] = y_train.values

train_df.to_csv("training_data.csv", index=False)


test_df = X_test.copy()
test_df["target"] = y_test.values

test_df.to_csv("testing_data.csv", index=False)
```
