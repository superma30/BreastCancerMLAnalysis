# Data Visualization

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import kagglehub
```

```python
os.environ['http_proxy'] = "http://172.18.0.1:3128"
os.environ['https_proxy'] = "http://172.18.0.1:3128"
```

```python
path = kagglehub.dataset_download("uciml/breast-cancer-wisconsin-data")
print("Path to dataset files:", path)

csv_path = os.path.join(path, "data.csv")
df = pd.read_csv(csv_path)

df["target"] = df["diagnosis"].map({"M": 0, "B": 1})
df = df.drop(columns=["id", "diagnosis", "Unnamed: 32"])

data = df.drop(columns=["target"])
labels = df["target"]
```

```python
# -----------------------------
# 1. Class Distribution
# -----------------------------
plt.figure(figsize=(6, 4))
sns.countplot(x='target', data=df)
plt.title('Class Distribution')
plt.xlabel('Diagnosis')
plt.ylabel('Count')
plt.show()
```

```python
# -----------------------------
# 2. Feature Histograms
# -----------------------------
df.hist(bins=20, figsize=(18, 12))
plt.tight_layout(pad=1.5)
plt.subplots_adjust(top=0.93)
plt.suptitle('Feature Distributions')
plt.show()
```

```python
# -----------------------------
# 3. Correlation Heatmap
# -----------------------------
plt.figure(figsize=(12, 10))
corr = df.drop(columns=['target']).corr()
sns.heatmap(corr, cmap='coolwarm', linewidths=0.5)
plt.title('Feature Correlation Heatmap', pad=20)
plt.show()
```

```python
# -----------------------------
# 4. Pairplot (subset of features)
# -----------------------------
selected_features = ['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean']
sns.pairplot(df[selected_features + ['target']], hue='target')
plt.show()
```

```python
# -----------------------------
# 4. Pairplot (all features)
# -----------------------------
selected_features = ['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean', 'compactness_mean',
                     'concavity_mean', 'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean', 'radius_se',
                     'texture_se', 'perimeter_se', 'area_se', 'smoothness_se', 'compactness_se', 'concavity_se',
                     'concave points_se', 'symmetry_se', 'fractal_dimension_se', 'radius_worst', 'texture_worst',
                     'perimeter_worst', 'area_worst', 'smoothness_worst', 'compactness_worst', 'concavity_worst',
                     'concave points_worst', 'symmetry_worst', 'fractal_dimension_worst']
sns.pairplot(df[selected_features + ['target']], hue='target')
plt.show()
```

```python
# -----------------------------
# 5. Boxplots (feature vs target)
# -----------------------------
plt.figure(figsize=(12, 6))
sns.boxplot(x='target', y='radius_mean', data=df)
plt.title('Mean Radius by Diagnosis')
plt.show()
```
