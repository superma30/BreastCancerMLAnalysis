### Download and basic processing
Here we:
- **download the dataset** from [kagglehub](https://www.kaggle.com/datasets/uciml/breast-cancer-wisconsin-data)
- get the **path of the data** and **load** it into a pandas **dataframe**
- **map the diagnosis** into a _"target"_ column with 0 as _Malignant_ and 1 as _Benign_
- **drop useless columns** and separate the _"target"_ column in a _labels_ dataframe

```python
path = kagglehub.dataset_download("uciml/breast-cancer-wisconsin-data")

csv_path = os.path.join(path, "data.csv")
df = pd.read_csv(csv_path)

df["target"] = df["diagnosis"].map({"M": 0, "B": 1})
df = df.drop(columns=["id", "diagnosis", "Unnamed: 32"])

data = df.drop(columns=["target"])
labels = df["target"]
```

## Class distribution
**Distribution** of the **_Benign_ (1)** and **_Malignant_ (0)** classes

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='target', data=df)
plt.title('Class Distribution')
plt.xlabel('Diagnosis')
plt.ylabel('Count')
plt.show()
```

## Feature Histograms
Distribution of all 31 features

```python
df.hist(bins=20, figsize=(18, 12))
plt.tight_layout(pad=1.5)
plt.subplots_adjust(top=0.93)
plt.suptitle('Feature Distributions')
plt.show()
```

## Correlation Heatmap
Heatmap of the correlation between all features

```python
plt.figure(figsize=(12, 10))
corr = df.drop(columns=['target']).corr()
sns.heatmap(corr, cmap='coolwarm', linewidths=0.5)
plt.title('Feature Correlation Heatmap', pad=20)
plt.show()
```

## Pairplot (subset)
Pairplot of a subset of the first 4 features between the two classes

```python
selected_features = ['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean']
sns.pairplot(df[selected_features + ['target']], hue='target')
plt.show()
```

## Boxplots
Boxplots of the mean radius related to the diagnosis

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='target', y='radius_mean', data=df)
plt.title('Mean Radius by Diagnosis')
plt.show()
```
