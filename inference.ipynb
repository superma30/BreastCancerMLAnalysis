# Inference
In this notebook we use the **test data** to **test** the **accuracy** of the previously trained **model**

```python
import pandas as pd
import sklearn
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
import joblib
import seaborn as sns
import matplotlib.pyplot as plt
```

## Loading
We load the **model** from `model.pkl` and the **testing data** from `testing_data.csv`

```python
classifier = joblib.load("model.pkl")

test_df = pd.read_csv("testing_data.csv")

X_test = test_df.drop(columns=["target"])
y_test = test_df["target"]
```

## Prediction
We **run inference** on the **testing data** and print the **accuracy score** and **classification report**

```python
# Predictions
y_pred = classifier.predict(X_test)

# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")

# Classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

## Confusion Matrix Heatmap
We make a **confusion matrix heatmap** which shows the **`True Positives`**, **`True Negatives`**, **`False Positives`** and **`False Negatives`** in the predictions

```python
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6, 5))

sns.heatmap(
    cm,
    annot=True,
    fmt='d',
    cmap='Blues',
    xticklabels=['Malignant', 'Benign'],
    yticklabels=['Malignant', 'Benign']
)

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")

plt.show()
```
