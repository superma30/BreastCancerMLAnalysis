# Training
In this notebook we train a simple **SVM** (**Support Vector Machine**s) model with the **previously pre-processed training data** and we **save the model**

```python
import pandas as pd
import sklearn
import joblib
```

## Load data and labels
We load the **previously pre-processed data** and split it into **X** and **Y** (**data** and **labels**)

```python
train_df = pd.read_csv("training_data.csv")

X_train = train_df.drop(columns=["target"]) # Contiene i dati (input)
y_train = train_df["target"] # Contiene le label (risultati da verificare)
```

## Best parameters search
Here we use a `GridSearchCV` to search the **best parameters** for the **SVC** (**Support Vector Classifier**)

```python
param_grid = { # Cerca i migliori parametri
    'C': [0.1, 1, 10, 100],
    'gamma': [1,0.1,0.01,0.001],
    'kernel' :['rbf']
}
grid = sklearn.model_selection.GridSearchCV(
    sklearn.svm.SVC(),
    param_grid,
    refit = True,
    verbose = 2,
    cv = 5,
)
grid.fit(X_train, y_train) # Qua prendiamo i migliori parametri per il modello
print(f"\nBest parameters: {grid.best_params_}")
print(f"\nBest estimator: {grid.best_estimator_}")
```

## Creating the model
We create the model from an *sklearn* `SVC` (**Support Vector Classifier**) **model** from the `svm` subcategory

```python
classificator = sklearn.svm.SVC(gamma=0.001,C=1., verbose=True)
# - C è il parametro di regolarizzazione, indica quanto è rigido il modello rispetto ai dati (misclassification)
# - Gamma controlla l'influenza di un punto, quanto è rilevante
```

## Training
We **fit** the model with the training data and print the **training score**

```python
classificator.fit(X_train, y_train)
score = classificator.score(X_train, y_train)
print("\nTraining score: "+str(score))
```

## Saving
We **save the model** to `model.pkl` with **joblib** (a library optimized for arrays of data)

```python
joblib.dump(classificator, "model.pkl")
```
