# Training
In this notebook we train a simple **SVM** (**Support Vector Machine**s) model with the **scaled** previously **pre-processed training data** and we **save the model**

```python
import pandas as pd
import sklearn
import joblib
```

## Load data and labels
We load the **previously pre-processed data** and split it into **X** and **Y** (**data** and **labels**)

```python
train_df = pd.read_csv("training_data.csv")

X_train = train_df.drop(columns=["target"]) # Contiene i dati (input)
y_train = train_df["target"] # Contiene le label (risultati da verificare)
```

## Pipeline creation
We create a `Pipeline` with a `StandardScaler` to **scale** the values by variance and an `SVC` (**Support Vector Classifier**)

```python

pipeline = sklearn.pipeline.Pipeline([
    ('scaler', sklearn.preprocessing.StandardScaler()),
    ('svc', sklearn.svm.SVC())
])
```

## Best parameters search
Here we use a `GridSearchCV` to search the **best parameters** for the **pipeline** and use the model with the best parameters

```python
param_grid = { # Cerca i migliori parametri
    'svc__C': [0.1, 1, 10, 100],
    'svc__gamma': [1, 0.1, 0.01, 0.001],
    'svc__kernel': ['rbf']
}

grid = sklearn.model_selection.GridSearchCV(
    pipeline,
    param_grid,
    refit = True,
    verbose = 2,
    cv = 5,
)

grid.fit(X_train, y_train) # Qua prendiamo i migliori parametri per il modello
print(f"\nBest parameters: {grid.best_params_}")
print(f"\nBest estimator: {grid.best_estimator_}")
```

## Creating the model
We get the model from the `best_estimator` of the Grid Search

```python
classifier = grid.best_estimator_
```

## Training
We **fit** the model with the training data and print the **training score**

```python
classifier.fit(X_train, y_train)
score = classifier.score(X_train, y_train)
print("\nTraining score: "+str(score))
```

## Saving
We **save the model** to `model.pkl` with **joblib** (a library optimized for arrays of data)

```python
joblib.dump(classifier, "model.pkl")
```
